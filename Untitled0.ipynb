```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Load the dataset
data = pd.read_csv('wine_quality.csv')

# Display the first few rows of the dataset
print(data.head())

# Check for missing values
print(data.isnull().sum())

# Impute missing values with mean
imputer = SimpleImputer(strategy='mean')
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

# Split data into input features (X) and output variable (y)
X = data_imputed.drop('quality', axis=1)
y = data_imputed['quality']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=data_imputed)
plt.title('Distribution of Wine Quality')
plt.show()

# Plot pairwise relationships
sns.pairplot(data_imputed, hue='quality')
plt.show()

# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data_imputed.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Select the top 5 features
selector = SelectKBest(f_classif, k=5)
X_new = selector.fit_transform(X_scaled, y)

# Get the column names of the selected features
selected_features = X.columns[selector.get_support()]

print("Selected features:", selected_features)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Convert the target variable to discrete values (if it's continuous)
y_train = y_train.astype(int)  # Convert to integers for classification
y_test = y_test.astype(int)    # Convert to integers for classification

# Train a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import IsolationForest

# Use Isolation Forest for outlier detection
iso_forest = IsolationForest(contamination=0.1, random_state=42)
outliers = iso_forest.fit_predict(X_scaled)

# Add the outlier information to the original data
data_imputed['outlier'] = outliers

# Plot the outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='volatile acidity', data=data_imputed, hue='outlier', palette='coolwarm')
plt.title('Outlier Detection')
plt.show()
```
